## To split strings on Uppercase characters

```python
# Python code to demonstrates to split strings on uppercase letters
import re

# initializing the string
init_str = str(input("Enter string"))

# To print the initial string
print("Initial String is:", init_str)

# Now splinting on uppercase using re method
res = re.findall('[A-Z][^A-Z]*', init_str)
print("Wanted string is:", str(res))
```

### One more way to split string

```python
str([s for s in res if s])
```

```python
' '.join(res)
```

```python
a = [i for i, e in enumerate(init_str+'A') if e.isupper()]
print(a)
```

```python
[init_str[a[j]:a[j +1]] for j in range(len(a)-1)]
```

```python
# Import the module
from geopy.geocoders import Nominatim
gl = Nominatim(user_agent = "geoapiExercises")
```

```python
# Longitude and Latitude
Long = "85.137566"
Lat = "25.594095"
location = gl.reverse(Lat+","+Long)
print(list(location))
```

```python
addr = location.raw['address']
print(addr)
```

```python
city = addr.get('city',' ')
state = addr.get('state',' ')
country = addr.get('country',' ')
country_code = addr.get('country_code',' ')
postcode = addr.get('postcode',' ')
print("city is:",city)
print("state name is:", state)
print("Country name is:", country)
print("Country Code is:", country_code)
print("Post Code is:", postcode)
```
